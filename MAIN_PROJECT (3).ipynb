### 5. Investigate if there's any correlation between age and salary, and represent it visually.

```python
correlation=df['Age'].corr(df['Salary'])
print(correlation)
```

```python
plt.title("Correlation b/n age and salary")
plt.xlabel("Salary")
plt.ylabel("Age")
x=df['Salary'].values
y=df['Age'].values
data=pd.DataFrame({'x':x,'y':y})
print(data)
sns.scatterplot(data=data,x='x',y='y')
plt.show()
```

#from this scatter plot of 'Age' with 'Salary' , correlation between age and salary are more clear.
By calculation we get correlation is 0.205,
